```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
import ipywidgets as widgets
from IPython.display import display

# Upload image
uploaded = files.upload()

# Load image using OpenCV
img = cv2.imread(list(uploaded.keys())[0])

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise and improve edge detection
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Function to update the Canny edge detection based on slider values
def update_edges(threshold1, threshold2):
    edges = cv2.Canny(blurred, threshold1, threshold2)
    cv2_imshow(edges)

# Sliders for controlling the thresholds
threshold1_slider = widgets.IntSlider(value=100, min=0, max=500, step=1, description='Threshold1')
threshold2_slider = widgets.IntSlider(value=200, min=0, max=500, step=1, description='Threshold2')

# Display the sliders and link them to the update function
output = widgets.interactive_output(update_edges, {'threshold1': threshold1_slider, 'threshold2': threshold2_slider})
display(threshold1_slider, threshold2_slider, output)

# Display the original image
#cv2_imshow(img)
```

```python
# Save the current edge-detected image
def save_image():
    edges = cv2.Canny(blurred, threshold1_slider.value, threshold2_slider.value)
    cv2.imwrite('edges_output.jpg', edges)
    files.download('edges_output.jpg')

# Button to download the image
download_button = widgets.Button(description="Download Image")
download_button.on_click(lambda x: save_image())
display(download_button)
```
